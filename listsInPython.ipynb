<h2>Change a Range of Item Values</h2>
To change the value of items within a specific range, define a list with the new values, and refer to the range of index numbers where you want to insert the new values:


```python
myStrList = ["apple","xaiomi","samsung","Vivo","Oppo","OnePlus","Nothing Phone"]

myStrList[2]="CHnaged"
print(myStrList)
```

```python
myStrList = ["apple","xaiomi","samsung","Vivo","Oppo","OnePlus","Nothing Phone"]

myStrList[1:3]=["index1","index2","index3"]
print(myStrList)

myStrList[2:]="cleared"
print(myStrList)

myStrList[2:]=["cleared"]
print(myStrList)
```

<h1>Insret Items in List | Note :</h1>

<h2>The insert() method inserts <i>An item</i> at the specified index. i.e.:only one item at once
    
    Where As extend()  is used to extend exisitng list by adding another list or other words.

```python
my_list = ["banana", "cherry", "orange"]

my_list.insert(0, "apple")#insert using insert()
print(my_list)

my_list = my_list + ["kela","sanatara"]#inserting us + operator
print(my_list)

addthis=["mango","licchi"]#insert using extend()
my_list.extend(addthis)
print(my_list)
```

<h3>#remove form list in python

```python
listIs = ['apple', 'banana', 'cherry', 'orange', 'kela', 'sanatara', 'mango', 'licchi']

x=listIs.pop(2)
print(x,"is poped")
print(listIs)

del listIs[2]
print(listIs,"\n. . .gone")
```
